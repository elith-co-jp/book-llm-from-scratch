# Attention の次元とスケーリングの関係

```python
import numpy as np
import matplotlib.pyplot as plt
import scienceplots
import japanize_matplotlib

plt.style.use(['science', 'japanese'])
```

```python
n_keys = 10000
dims = [10, 100, 200, 500, 1000]
vars = []
for dim in dims:
    query = np.random.normal(size=(dim,))
    keys = np.random.normal(size=(n_keys, dim))

    weight = np.dot(keys, query) 
    variance = np.var(weight)
    vars.append(variance)
```

```python
fig, ax = plt.subplots(figsize=(4, 3))
ax.set_xlabel('次元数')
ax.set_ylabel('分散')
ax.plot(dims, vars, marker='o')
plt.savefig('scaled_attention.png', bbox_inches='tight', dpi=300)
```

```python
n_keys = 10000
dims = [10, 100, 200, 500, 1000]
vars = []
for dim in dims:
    query = np.random.normal(size=(dim,))
    keys = np.random.normal(size=(n_keys, dim))

    weight = np.dot(keys, query) / np.sqrt(dim)
    variance = np.var(weight)
    vars.append(variance)
```

```python
fig, ax = plt.subplots(figsize=(4, 3))
ax.set_xlabel('次元数')
ax.set_ylabel('分散')
ax.set_ylim(0, 1.5)
ax.plot(dims, vars, marker='o')
plt.savefig('scaled_attention_weight.png', bbox_inches='tight', dpi=300)
```
